# **Minesweeper Game**

### We are simulating a Minesweeper game using Object-Orienting Programming (OOP) in Python. For that, we are going to creeate a class with different methods until we have a complete game

## **STEP 1: GAMEBOARD CREATION**

### We are defining the class and create a method to generate the gameboard. Parameters of this method are the number of rows, columns and the character which we want to fill the gameboard with. This method will return the same number of list as rows. These lists will have the same lenth as the number of colums

```python
class Minesweeper:

  def initial_gameboard(self, rows, columns, character):
    self.rows = rows
    self.columns = columns
    self.character = character
    gameboard = []

    for i in range(rows):
      i = [character] * columns 
      gameboard.append(i)
      
    return gameboard




Minesweeper().initial_gameboard(5,5,0)


```

## **STEP 2: ADD MINES IN THE GAMEBOARD**

### Now, we are creating the method *add_mines**. This method will add randomly mines in the gameboard. Parameters will be the number of rows, columns and the number of mines. 

### To create this method, we should have in mind that it have to:
- Call the previous method (*initial_gameboard*)
- Check the number of mines is smaller than the number of cells in the gameboard. In the case of the number would be higher, a warning message would be printed.
- Add randomly mines in the gameboard, replacing 0s wiyh Ms. Check that the final number of mines is equal that the number introduced as a parameter 

```python
import random

class Minesweeper:

  def initial_gameboard(self, rows, columns, character):
    self.rows = rows
    self.columns = columns
    self.character = character
    gameboard = []

    for i in range(rows):
      i = [character] * columns 
      gameboard.append(i)
      
    return gameboard

  def add_mines (self, rows, columns, mines):
    self.rows = rows
    self.columns = columns
    self.mines = mines
    #Call the previous method
    gameboard = self.initial_gameboard(rows, columns, 0)
    #Add mines

    if mines < rows*columns:
      #Create a loop for adding randomly mines

      for mine in range(mines):
        row = random.randint(0, rows - 1) 
        column = random.randint(0, columns - 1)

        #In case that in the cell would be a M
        while gameboard [row][column] == "M": 
          row = random.randint(0, row - 1)
          column = random.randint(0, columns - 1)
        gameboard[row][column] = "M" 

      return gameboard
    
    else:
      print ("The number of mines cannot be higher than the number of cells. Write a lower number")


Minesweeper().add_mines(5,5,5)

```

## **STEP 3: GENERATING CLUES IN THE GAMEBOARD**

### Now, clues that indicates how many mines are around the cell are adding. We are creating a new method which has as parameters the number of rows, columns and mines. For that, it has to:
- Call the method *add_mines* and create the gameboard with mines
- Look for mines and sum 1 in the adjacent cells
- Print the resulting list

```python
import random

class Minesweeper:

  def initial_gameboard(self, rows, columns, character):
    self.rows = rows
    self.columns = columns
    self.character = character
    gameboard = []

    for i in range(rows):
      i = [character] * columns 
      gameboard.append(i)
      
    return gameboard

  def add_mines (self, rows, columns, mines):
    self.rows = rows
    self.columns = columns
    self.mines = mines
    gameboard = self.initial_gameboard(rows, columns, 0)
    
    if mines < rows*columns:

      for mine in range(mines):
        row = random.randint(0, rows - 1) 
        column = random.randint(0, columns - 1)
        
        while gameboard [row][column] == "M": 
          row = random.randint(0, row - 1)
          column = random.randint(0, columns - 1)
        gameboard[row][column] = "M" 
      return gameboard
    else:
      print ("The number of mines cannot be higher than the number of cells. Write a lower number")

  def numbers(self, rows, columns, mines):
    self.rows = rows
    self.columns = columns
    self.mines = mines
    #Add mines
    mines_gameboard = self.add_mines(rows, columns, mines)

    #Now keys are added with the numbers of adjacent mines
    for i in range(0,self.rows): 

      for j in range(0,self.columns): 
        #If there is mine in the position:

        if mines_gameboard[i][j]=='M':
          #Sum 1 in the row below (if it not the first row and there is not a mine) 
          if i!=0 and mines_gameboard[i-1][j]!='M':
            mines_gameboard[i-1][j]+=1
          #Sum 1 in the next row
          if i!=(rows-1) and mines_gameboard[i+1][j]!='M':
            mines_gameboard[i+1][j]+=1
          #Left column
          if j!=0 and mines_gameboard[i][j-1]!='M':
            mines_gameboard[i][j-1]+=1
          #Right column
          if j!=(columns-1) and mines_gameboard[i][j+1]!='M':
            mines_gameboard[i][j+1]+=1
          #Diagonal
          #Up-left
          if i!=0 and j!=0 and mines_gameboard[i-1][j-1]!='M':
            mines_gameboard[i-1][j-1]+=1
          #Up-right
          if i!=0 and j!=(columns-1) and mines_gameboard[i-1][j+1]!='M':
            mines_gameboard[i-1][j+1]+=1
          if i!=(rows-1) and j!=(columns-1) and mines_gameboard[i+1][j+1]!='M':
            mines_gameboard[i+1][j+1]+=1
          if i!=rows-1 and j!=0 and mines_gameboard[i+1][j-1]!='M':
            mines_gameboard[i+1][j-1]+=1 
            
    return mines_gameboard



Minesweeper().numbers(5,5,5)
```

## **STEP 4: PLAY!!**

### At last, we can play. User will be asked for a position. Code checks the position and show the board until a mine appears. Remember that the position in the first column and row is 0, 0

```python
import random

class Minesweeper:

  def initial_gameboard(self, rows, columns, character):
    self.rows = rows
    self.columns = columns
    self.character = character
    gameboard = []

    for i in range(rows):
      i = [character] * columns 
      gameboard.append(i)

    return gameboard

  def add_mines (self, rows, columns, mines):
    self.rows = rows
    self.columns = columns
    self.mines = mines
    gameboard = self.initial_gameboard(rows, columns, 0)
   
    if mines < rows*columns:

      for mine in range(mines):
        row = random.randint(0, rows - 1) 
        column = random.randint(0, columns - 1)
        
        while gameboard [row][column] == "M": 
          row = random.randint(0, row - 1)
          column = random.randint(0, columns - 1)
        gameboard[row][column] = "M" 

      return gameboard
    
    else:
      print ("The number of mines cannot be higher than the number of cells. Write a lower number")

  def numbers(self, rows, columns, mines):
    self.rows = rows
    self.columns = columns
    self.mines = mines
    mines_gameboard = self.add_mines(rows, columns, mines)
    
    for i in range(0,self.rows): 

      for j in range(0,self.columns): 
        
        if mines_gameboard[i][j]=='M':
          if i!=0 and mines_gameboard[i-1][j]!='M':
            mines_gameboard[i-1][j]+=1
          if i!=(rows-1) and mines_gameboard[i+1][j]!='M':
            mines_gameboard[i+1][j]+=1
          if j!=0 and mines_gameboard[i][j-1]!='M':
            mines_gameboard[i][j-1]+=1
          if j!=(columns-1) and mines_gameboard[i][j+1]!='M':
            mines_gameboard[i][j+1]+=1
          if i!=0 and j!=0 and mines_gameboard[i-1][j-1]!='M':
            mines_gameboard[i-1][j-1]+=1
          if i!=0 and j!=(columns-1) and mines_gameboard[i-1][j+1]!='M':
            mines_gameboard[i-1][j+1]+=1
          if i!=(rows-1) and j!=(columns-1) and mines_gameboard[i+1][j+1]!='M':
            mines_gameboard[i+1][j+1]+=1
          if i!=rows-1 and j!=0 and mines_gameboard[i+1][j-1]!='M':
            mines_gameboard[i+1][j-1]+=1 
    return mines_gameboard
  
  def play(self, rows, columns, mines):
    self.rows = rows
    self.columns = columns
    self.mines = mines
    # Creating the gameboard with solution
    gameboard_solution = self.numbers(rows, columns, mines)
    # Creating the initial gameboard
    gameboard = self.initial_gameboard(rows, columns, " ")

    # Ask the user for a position (rows, columns). Loop until the user's position is a mine.
    revealed_cells = 0
    game_over = False
    while True:
      try:
        user_position = input("Introduce a row and column (row, column). Keep in mind that the first row and the first column are 0,0: ")
        user_position_list = user_position.split(",")
        row = int(user_position_list[0])
        column = int(user_position_list[1])
        solution = gameboard_solution[row][column]

        if solution != "M" and gameboard[row][column] == " ":
          gameboard[row][column] = solution
          revealed_cells += 1
          for row in gameboard:
            print(row)

          if revealed_cells == rows * columns - mines:
            print("Congratulations! You've won!")
            break

        elif solution == "M":
          game_over = True
          break

        else:
          print("Position already revealed. Try again.")

      except IndexError:
          print("Position out of range")

      except ValueError:
          print("Incorrect format")

    if game_over:
        print("Game Over")
        # Reveal all mine locations
        for i in range(rows):
            for j in range(columns):
                if gameboard_solution[i][j] == "M":
                    gameboard[i][j] = "M"

        for row in gameboard:
            print(row)



Minesweeper().play(5,5,5)
```
