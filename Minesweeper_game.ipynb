# **Minesweeper Game**

## We are simulating a Minesweeper game using Object-Orienting Programming (OOP) in Python. For that, we are going to creeate a class with different methods until we have a complete game

```python
import random

class Minesweeper:

  def initial_gameboard(self, rows, columns, character):
    self.rows = rows
    self.columns = columns
    self.character = character
    gameboard = []

    for i in range(rows):
      i = [character] * columns 
      gameboard.append(i)

    return gameboard

  def add_mines (self, rows, columns, mines):
    self.rows = rows
    self.columns = columns
    self.mines = mines
    gameboard = self.initial_gameboard(rows, columns, 0)
   
    if mines < rows*columns:

      for mine in range(mines):
        row = random.randint(0, rows - 1) 
        column = random.randint(0, columns - 1)
        
        while gameboard [row][column] == "M": 
          row = random.randint(0, row - 1)
          column = random.randint(0, columns - 1)
        gameboard[row][column] = "M" 

      return gameboard
    
    else:
      print ("The number of mines cannot be higher than the number of cells. Write a lower number")

  def numbers(self, rows, columns, mines):
    self.rows = rows
    self.columns = columns
    self.mines = mines
    mines_gameboard = self.add_mines(rows, columns, mines)
    
    for i in range(0,self.rows): 

      for j in range(0,self.columns): 
        
        if mines_gameboard[i][j]=='M':
          if i!=0 and mines_gameboard[i-1][j]!='M':
            mines_gameboard[i-1][j]+=1
          if i!=(rows-1) and mines_gameboard[i+1][j]!='M':
            mines_gameboard[i+1][j]+=1
          if j!=0 and mines_gameboard[i][j-1]!='M':
            mines_gameboard[i][j-1]+=1
          if j!=(columns-1) and mines_gameboard[i][j+1]!='M':
            mines_gameboard[i][j+1]+=1
          if i!=0 and j!=0 and mines_gameboard[i-1][j-1]!='M':
            mines_gameboard[i-1][j-1]+=1
          if i!=0 and j!=(columns-1) and mines_gameboard[i-1][j+1]!='M':
            mines_gameboard[i-1][j+1]+=1
          if i!=(rows-1) and j!=(columns-1) and mines_gameboard[i+1][j+1]!='M':
            mines_gameboard[i+1][j+1]+=1
          if i!=rows-1 and j!=0 and mines_gameboard[i+1][j-1]!='M':
            mines_gameboard[i+1][j-1]+=1 
    return mines_gameboard
  
  def play(self, rows, columns, mines):
    self.rows = rows
    self.columns = columns
    self.mines = mines
    gameboard_solution = self.numbers(rows, columns, mines)
    gameboard = self.initial_gameboard(rows, columns, " ")

    revealed_cells = 0
    game_over = False
    while True:
      try:
        user_position = input(
          "Introduce a row and column (row, column). Keep in mind that the first row and the first column are 0,0: ")
        user_position_list = user_position.split(",")
        row = int(user_position_list[0])
        column = int(user_position_list[1])
        solution = gameboard_solution[row][column]

        if solution != "M" and gameboard[row][column] == " ":
          gameboard[row][column] = solution
          revealed_cells += 1
          for row in gameboard:
            print(row)

          if revealed_cells == rows * columns - mines:
            print("Congratulations! You've won!")
            break

        elif solution == "M":
          game_over = True
          break

        else:
          print("Position already revealed. Try again.")

      except IndexError:
          print("Position out of range")

      except ValueError:
          print("Incorrect format")

    if game_over:
        print("Game Over")
        for i in range(rows):
            for j in range(columns):
                if gameboard_solution[i][j] == "M":
                    gameboard[i][j] = "M"

        for row in gameboard:
            print(row)



Minesweeper().play(5,5,5)
```
